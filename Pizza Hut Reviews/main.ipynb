```python
from google.colab import drive
drive.mount('/content/drive')

%cd "/content/drive/MyDrive/Colab Notebooks/Python for Data Science/Project"

!jupyter nbconvert --to python data_loader.ipynb
!jupyter nbconvert --to python sentiment_analyzer.ipynb
!jupyter nbconvert --to python visualizer.ipynb
import nltk
nltk.download('stopwords')


# main.py

from data_loader import load_and_clean_data
from sentiment_analyzer import analyze_sentiment
from visualizer import create_visualizations

def main():
    # Load and clean data
    df = load_and_clean_data('pizza_hut_reviews.csv')
    if df is None:
        return

    # Classify sentiment
    df = analyze_sentiment(df)
    if df is None:
        return

    # Generate visualizations
    create_visualizations(df)

    # Save final labeled dataset
    try:
        df.to_csv('sentiment_report.csv', index=False)
    except Exception as e:
        print(f"Error saving sentiment_report.csv: {e}")

if __name__ == "__main__":
    main()
```
