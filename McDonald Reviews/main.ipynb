```python
# main.py
from google.colab import drive
drive.mount('/content/drive')

%cd "/content/drive/MyDrive/Colab Notebooks/Python for Data Science/Project2"

!jupyter nbconvert --to python data_loader.ipynb
!jupyter nbconvert --to python sentiment_analyzer.ipynb
!jupyter nbconvert --to python visualizer.ipynb
import nltk
nltk.download('stopwords')

from data_loader import load_data
from sentiment_analyzer import classify_sentiment
from visualizer import plot_sentiment_distribution, plot_sentiment_trend
import pandas as pd

def main():
    # Step 1: Load data
    file_path = 'McDonald_s_Reviews.csv'  # adjust if needed
    df = load_data(file_path)
    if df is None:
        return

    # Step 2: Classify sentiment using BERT
    print("Classifying sentiment... this may take a while")
    df['sentiment'] = df['review'].apply(classify_sentiment)

    # Step 3: Save intermediate results
    try:
        df.to_csv('sentiment_report.csv', index=False)
        print("Sentiment report saved as sentiment_report.csv")
    except Exception as e:
        print(f"Could not save CSV: {e}")

    # Step 4: Sentiment distribution visualization
    sentiment_counts = df['sentiment'].value_counts().to_dict()
    for cat in ['positive','neutral','negative']:
        sentiment_counts.setdefault(cat, 0)
    plot_sentiment_distribution(sentiment_counts, output_file='sentiment_distribution.png')
    print("Sentiment distribution chart saved as sentiment_distribution.png")

    # Step 5: Sentiment trend visualization
    if df['date'].notna().any():
        plot_sentiment_trend(df, output_file='sentiment_trend.png')
        print("Sentiment trend chart saved as sentiment_trend.png")
    else:
        print("No timestamp data found — trend chart skipped.")

if __name__ == "__main__":
    main()
```
